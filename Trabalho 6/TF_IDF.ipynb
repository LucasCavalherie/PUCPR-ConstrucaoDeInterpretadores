```python
# Lucas Fernando Assunção Cavalherie

# Sua tarefa será gerar a matriz termo-documento usando TF-IDF por meio da aplicação das 
# fórmulas  TF-IDF  na  matriz  termo-documento  criada  com  a  utilização  do  algoritmo  Bag of 
# Words. Sobre o Corpus que recuperamos anteriormente. O entregável desta tarefa é uma 
# matriz termo-documento onde a primeira linha são os termos e as linhas subsequentes são 
# os vetores calculados com o TF-IDF. 

# 2. Sua tarefa será gerar uma matriz de distância, computando o cosseno do ângulo entre todos 
# os vetores que encontramos usando o tf-idf. Para isso use a seguinte fórmula para o cálculo 
# do  cosseno  use  a  fórmula  apresentada  em  Word2Vector  (frankalcantara.com) 
# (https://frankalcantara.com/Aulas/Nlp/out/Aula4.html#/0/4/2)  e  apresentada  na  figura  a 
# seguir:  

from tabulate import tabulate
from bs4 import BeautifulSoup
import requests
import string
import re
import numpy as np

vocabulario = set()
palavras = []
textoPorUrl = []
urls = []

urls.append('https://developer.apple.com/documentation/naturallanguage')
urls.append('https://en.wikipedia.org/wiki/Natural_language_processing')
urls.append('https://www.ibm.com/cloud/learn/natural-language-processing')
urls.append('https://www.techtarget.com/searchenterpriseai/definition/natural-language-processing-NLP')
urls.append('https://www.sas.com/en_us/insights/analytics/what-is-natural-language-processing-nlp.html')

for url in urls:
  html = requests.get(url)
  texto = BeautifulSoup(html.text, "html.parser")

  for sentenca in texto.find_all("p"):
    sentenca = sentenca.get_text().translate(str.maketrans('', '', string.punctuation)) # remove pontuacao
    textoPorUrl.append(sentenca)

for texto in textoPorUrl:
  palavras.append(texto.split())

texto = palavras

for sentenca in texto:
  for palavra in sentenca:
    vocabulario.add(palavra)



print("BAG OF WORDS: ")
matriz = []
for sentenca in texto:
  vetor = []
  for palavra in vocabulario:
    if palavra in sentenca:
      vetor.append(sentenca.count(palavra))
    else:
      vetor.append(0)
  matriz.append(vetor)
#print(tabulate(matriz, headers=vocabulario))



print("TF: ")
matrizTF = []
for sentenca in texto:
  vetor = []
  tamSentenca = len(sentenca)
  for palavra in vocabulario:
    if palavra in sentenca:
      vetor.append(sentenca.count(palavra)/tamSentenca)
    else:
      vetor.append(0)
  matrizTF.append(vetor)
#print(tabulate(matrizTF, headers=vocabulario))



print("IDF: ")
matrizIDF = []
vetor = []
nTextos = len(texto)
for palavra in vocabulario:
  contador = 0
  for sentenca in texto:
    if palavra in sentenca:
      contador = contador + 1
  vetor.append(np.log(nTextos/contador))
matrizIDF.append(vetor)
#print(tabulate(matrizIDF, headers=vocabulario))



print("TF-IDF: ")
matrizTF_IDF = []
for linhaTF in matrizTF:
  vetorTF_IDF = []
  for i in range(len(linhaTF)):
    vetorTF_IDF.append(linhaTF[i])
  matrizTF_IDF.append(vetorTF_IDF)
print(tabulate(matrizTF_IDF, headers=vocabulario))



print("COSSINE SIMILARITY: ")
matrizCOSSINE = []
header = [] 
i = 1
for a in matrizTF:
  header.append("Vetor" + str(i))
  vetorCOSSINE = []
  for b in matrizTF:
    cos_sim = np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
    vetorCOSSINE.append(cos_sim)
  matrizCOSSINE.append(vetorCOSSINE)
  i = i + 1

print(tabulate(matrizCOSSINE, headers=header))

```
